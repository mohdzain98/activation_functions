```python
def relu(x):
    if(x>=0):
        return x
    else:
        return 0
```

```python
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-1))/(np.exp(x)+np.exp(-x))
```

```python
def leaky_relu(x):
    if(x>0):
        return x
    else:
        return x*0.01
```

```python
def softmax(x):
    z=exp(x)
    return z/z.sum()
    
```

```python
from scipy.special import softmax
data=[i for i in range(-10,10)]
result=softmax(data)
```

```python
def binarystep(x):
    if(x<0):
        return 0
    else:
        return 1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
plotsize()
x=[i for i in range(-10,10)]
y=[relu(i) for i in range(-10,10)]
z=[sigmoid(i) for i in range(-10,10)]
tan=[tanh(i) for i in range(-9,9)]
prelu=[leaky_relu(i) for i in range(-9,9)]
xx=[i for i in range(-9,9)]
softm=[softmax(i) for i in range(-10,10)]
binary=[binarystep(i) for i in range(-9,9)]
sns.lineplot(x=x,y=y,label='relu')
sns.lineplot(x=xx,y=prelu,label='leaky_relu(-9,9)')
sns.lineplot(x=x,y=result,label='softmax')
sns.lineplot(x=xx,y=tan,label='tanh(-9,9)')
sns.lineplot(x=x,y=z,label='sigmoid')
sns.lineplot(x=xx,y=binary,label='binary step(-9,9)')
```

```python
def hardtanh(x):
    if(x>=-1 and x<=1):
        return x
    elif(x>1):
        return 1
    else:
        return -1
```

```python
def hardsigmoid(x):
    return max(0,min(1,(x+1)/2))
```

```python
def swish(beta,x):
    return x*sigmoid(beta*x)
```

```python
def silu(beta,x):
    return x*sigmoid(beta*x)
```

```python
def plotsize():
    f = plt.figure()
    f.set_figwidth(9)
    f.set_figheight(6)
    plt.grid()
```

```python
x=[i for i in range(-10,10)]
htan=[hardtanh(i) for i in range(-10,10)]
hsig=[hardsigmoid(i) for i in range(-10,10)]
sil=[silu(1,i) for i in range(-10,10)]
```
